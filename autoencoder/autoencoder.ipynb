```python
import pandas as pd
import numpy as np
from sklearn import metrics
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
```

```python
df = pd.read_csv('data_ret.csv')
```

```python
x = df.RET_MSISDN.unique()
```

```python
x
```

```python
dic ={}
c = -1
for i in x:
    c=c+1
    dic[i]=c
    
```

```python
df['RET_MSISDN'] = df['RET_MSISDN'].map(dic)
```

```python
df
```

```python
value_counts = df['RET_MSISDN'].value_counts()
```

```python
value_counts.max
```

```python
train = df.values
```

```python
model = Sequential()
model.add(Dense(50,input_dim=train.shape[1],activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(train.shape[1]))
model.compile(loss='mse',optimizer='adam')
model.fit(train,train,verbose=1,epochs=100)
```

```python
pred = model.predict(train)
score = np.sqrt(metrics.mean_squared_error(pred,train))
print(f"Normal Score : {score}".format(score))
```
